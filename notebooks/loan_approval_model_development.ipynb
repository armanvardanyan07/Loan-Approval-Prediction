```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_train = df_train.drop('id',axis=1)
df_test = df_test.drop('id',axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
df_train['person_home_ownership'] = label.fit_transform(df_train['person_home_ownership'])
df_test['person_home_ownership'] = label.fit_transform(df_test['person_home_ownership'])
```

```python
df_train['loan_intent'] = label.fit_transform(df_train['loan_intent'])
df_test['loan_intent'] = label.fit_transform(df_test['loan_intent'])
```

```python
df_train['loan_grade'] = label.fit_transform(df_train['loan_grade'])
df_test['loan_grade'] = label.fit_transform(df_test['loan_grade'])
```

```python
df_train['cb_person_default_on_file'] = label.fit_transform(df_train['cb_person_default_on_file'])
df_test['cb_person_default_on_file'] = label.fit_transform(df_test['cb_person_default_on_file'])
```

```python
X = df_train.drop('loan_status',axis=1)
y = df_train['loan_status']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
X_test = df_test.copy()
```

```python
import pandas as pd
from catboost import CatBoostClassifier, Pool
feature_cols = [col for col in X.columns if col != 'id'] 
X_train = X[feature_cols]
X_test = df_test[feature_cols]
cat_features = [
    'person_home_ownership',
    'loan_intent',
    'loan_grade',
    'cb_person_default_on_file'
]
train_pool = Pool(X_train, y, cat_features=cat_features)
test_pool = Pool(X_test, cat_features=cat_features)
model = CatBoostClassifier(
    iterations=200,
    verbose=50,
    max_ctr_complexity=1,
)
model.fit(train_pool)
y_pred = model.predict_proba(test_pool)[:, 1]
df_sub['loan_status'] = y_pred
df_sub.to_csv('vark.csv',index=False)
```
